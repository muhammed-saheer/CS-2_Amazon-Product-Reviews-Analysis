```python
import numpy as np

# Function to clean numeric columns (₹, %, ,)
def clean_numeric_column(series):
    return pd.to_numeric(series.replace('[₹%,]', '', regex=True).replace('', np.nan))
```

```python
# Apply cleaning
df['discounted_price'] = clean_numeric_column(df['discounted_price'])
df['actual_price'] = clean_numeric_column(df['actual_price'])
df['discount_percentage'] = clean_numeric_column(df['discount_percentage'])
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')
```

# Q1: How does the discount percentage affect the rating of a product?

In this analysis, we want to check if higher discounts result in better or worse product ratings.

We will:
- Convert discount and rating columns to numeric types (already done in cleaning).
- Visualize the relationship using a scatter plot.
- Calculate the correlation coefficient between discount percentage and rating.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='discount_percentage', y='rating', data=df, alpha=0.6)
plt.title('Discount % vs Product Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Product Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Correlation
discount_rating_corr = df['discount_percentage'].corr(df['rating'])
discount_rating_corr
```

### Observation:

The scatter plot shows a scattered distribution with no strong linear pattern.

The correlation coefficient is ~ -0.15, which means a very weak negative correlation.

### Conclusion:

Products with higher discounts tend to have slightly lower ratings, but the effect is very weak. So, discounts don't significantly influence customer ratings.

- Group by `main_category` and calculate the average rating

# Q3: Is there a correlation between the product's price and its rating?


In this analysis, we aim to determine whether **more expensive products receive better ratings**.

📌 **Steps:**
- Use the `discounted_price` and `rating` columns from the cleaned dataset.
- Plot a **scatter plot** to visually observe any trend between price and rating.
- Use the `.corr()` function to calculate the **Pearson correlation coefficient** to measure the strength and direction of the relationship.

🔎 **Interpretation Guide:**
- A correlation near **0** implies no linear relationship.
- A **positive correlation** (closer to +1) means higher price = higher rating.
- A **negative correlation** (closer to -1) means higher price = lower rating.

```python
# Scatter plot to visualize relationship
plt.figure(figsize=(8, 5))
sns.scatterplot(x='discounted_price', y='rating', data=df, alpha=0.6)
plt.title('Price vs Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate and display correlation
price_rating_corr = df['discounted_price'].corr(df['rating'])
print(f"Correlation between product price and rating: {price_rating_corr:.3f}")
```

### Observation Based on the Plot:
- The plot shows that most products are priced under ₹10,000, and their ratings are mostly between 4.0 and 4.5.
- Even products with very high prices (₹50K+) do not necessarily get higher ratings.
- There is a dense cluster of affordable products with good ratings, indicating that lower price doesn't mean lower satisfaction.

A few low-rated outliers exist across all price points, showing variability in perceived value.

### Conclusion:
Although the correlation between price and rating is slightly positive (0.120), the scatter plot reveals no strong trend. This means price is not a key driver of customer satisfaction. Customers likely base their ratings more on factors such as quality, expectations, brand, and experience rather than price alone.



# Q4: What is the most common word in the positive and negative reviews?

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import string

# Download stopwords (if not already)
nltk.download('stopwords')

# Combine all review content into a single string for each sentiment group
positive_reviews = df[df['rating'] >= 4]['review_content'].dropna().str.lower().str.cat(sep=' ')
negative_reviews = df[df['rating'] < 4]['review_content'].dropna().str.lower().str.cat(sep=' ')

# Function to clean and tokenize text
def get_clean_tokens(text):
    tokens = text.split()
    stop_words = set(stopwords.words('english'))
    return [word.strip(string.punctuation) for word in tokens if word not in stop_words and word.isalpha()]

# Get clean tokens
positive_tokens = get_clean_tokens(positive_reviews)
negative_tokens = get_clean_tokens(negative_reviews)

# Count word frequencies
positive_freq = Counter(positive_tokens).most_common(10)
negative_freq = Counter(negative_tokens).most_common(10)

print("Top Positive Words:", positive_freq)
print("Top Negative Words:", negative_freq)
```

```python
# Convert to DataFrame for plotting
import pandas as pd
import matplotlib.pyplot as plt

pos_df = pd.DataFrame(positive_freq, columns=['word', 'count'])
neg_df = pd.DataFrame(negative_freq, columns=['word', 'count'])

# Plot Positive
plt.figure(figsize=(8,4))
plt.bar(pos_df['word'], pos_df['count'], color='green')
plt.title("Top Words in Positive Reviews")
plt.show()

# Plot Negative
plt.figure(figsize=(8,4))
plt.bar(neg_df['word'], neg_df['count'], color='red')
plt.title("Top Words in Negative Reviews")
plt.show()
```

### Observation:
- Surprisingly, “good” appears in both positive and negative reviews. This shows that even negative reviews may contain some praise, or the word "good" might be followed by a "but..."
- Words like “battery” and “working” are specific to negative reviews, indicating functionality issues are common complaints.
- In positive reviews, terms like “quality”, “cable”, and “using” suggest satisfaction with build and performance.

### Conclusion:
Review text shows that customers consistently mention “good product quality” in positive reviews. However, functionality-related concerns (e.g., "battery", "working") are more common in negative feedback. These insights can help businesses focus on improving product reliability and highlighting key features in marketing content.

# Q6: Most Reviewed Product & Its Rating

In this question, we aim to identify the **most reviewed product**, as it likely represents a **high-selling or widely used item**.

📌 **Steps:**
- Group the dataset by `product_id` and `product_name`
- Sum the `rating_count` for each product
- Use `.idxmax()` to find the product with the highest number of reviews
- Retrieve and display its rating and other details

```python
# Group by product_id and product_name, and sum the rating_count
review_counts = df.groupby(['product_id', 'product_name']).agg({
    'rating_count': 'sum',
    'rating': 'mean'
}).reset_index()

# Find the product with the highest number of reviews
most_reviewed = review_counts.loc[review_counts['rating_count'].idxmax()]
most_reviewed
```

### Observation:
The AmazonBasics HDMI Cable has the highest number of reviews (853,945) — far more than any other product in the dataset.

Despite the high review count, it maintains a strong average rating of 4.4, which speaks to both its popularity and customer satisfaction.

### Conclusion:
The most reviewed product is the AmazonBasics Flexible HDMI Cable, with over 850k reviews and a solid rating of 4.4. This indicates the product is likely high-selling, reliable, and enjoys strong word-of-mouth. Such products can be used as benchmarks for customer satisfaction or leveraged in marketing efforts (e.g., “#1 rated cable”).

# Q7: Top 5 Users by Number of Reviews

In this step, we aim to find the most active reviewers. These users are valuable for:
- Product feedback
- Loyalty programs
- Influencer targeting

📌 **Steps:**
- Split the `user_id` column (comma-separated) into individual IDs
- Count frequency of each user
- Sort to find the top 5

```python
from collections import Counter

# Split all user IDs (comma-separated) and flatten into one list
all_user_ids = df['user_id'].dropna().astype(str).str.split(',').sum()

# Count frequency of each user_id
user_freq = Counter(all_user_ids)

# Get top 5 users
top_users = user_freq.most_common(5)
top_users
```

### Observation:

- No “power reviewers” exist in this dataset (e.g., hundreds of reviews per user).

### Conclusion:
The top 5 reviewers have posted around 10–11 reviews each. These may be engaged or returning customers, making them good candidates for personalized campaigns, feedback surveys, or product testing programs. However, the relatively low count suggests that most users are casual reviewers.

# Q8: Correlation Between Review Length and Rating

Here we analyze whether **longer reviews tend to be more positive or more negative**.

📌 **Steps:**
- Measure the length of each review (number of characters or words)
- Create a new column `review_length`
- Use `.corr()` to find the correlation between `review_length` and `rating`
- Visualize the relationship using a scatter plot

```python
# Create a new column for review length (number of characters)
df['review_length'] = df['review_content'].fillna('').astype(str).apply(len)

# Calculate correlation
length_rating_corr = df['review_length'].corr(df['rating'])
print(f"Correlation between review length and rating: {length_rating_corr:.3f}")
```

### Observation:
- The correlation coefficient is 0.077, indicating a very weak positive relationship.
- The scatter plot shows no strong trend — both short and long reviews exist across all rating levels.
- Some long reviews do appear with both high and low ratings, likely when users feel strongly (positively or negatively) about a product.

### Conclusion:
There is no significant correlation between the length of a review and the rating given. While slightly longer reviews tend to be mildly more positive, the effect is negligible. This suggests that review length alone is not a reliable indicator of customer sentiment — further text sentiment analysis would provide deeper insights.

# Q9: Does having an image link affect the product rating?

In this question, we want to see if products that include an image link tend to receive **higher or lower ratings** compared to those that don’t.

📌 **Steps:**
- Create a new column `has_image` → True if image link exists, else False
- Group by this column and compare average ratings
- Visualize the result with a bar chart

```python
# Create a new column: has_image (True if image link is not empty or NaN)
df['has_image'] = df['img_link'].notnull() & df['img_link'].astype(str).str.startswith('http')

# Group by has_image and calculate average rating
image_rating = df.groupby('has_image')['rating'].mean()
print(image_rating)
```

```python
import matplotlib.pyplot as plt

# Plot bar chart
image_rating.plot(kind='bar', color=['gray', 'orange'], figsize=(6,4))
plt.title('Average Rating: Products with vs without Image')
plt.xlabel('Has Image?')
plt.ylabel('Average Rating')
plt.xticks(ticks=[0,1], labels=['No', 'Yes'], rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Observation 
- All products in the dataset have a valid img_link, so there’s no comparison group to evaluate the impact of missing images.
- The average rating for products with images is 4.10, suggesting a good overall satisfaction level.
- Since there are no products without images, we cannot determine the effect of image presence on rating.

### Conclusion:
In this dataset, all products have image links, so we can’t compare ratings with and without images. However, the average rating of 4.10 for products with images may suggest that having visuals helps customers feel more confident, but we can’t confirm causality without a comparison group.

# Q10: Correlation Between Product Description Length and Rating

In this analysis, we examine whether **longer product descriptions lead to better customer ratings**.

📌 **Steps:**
- Calculate the length of each product description (`about_product` column)
- Create a new column `description_length`
- Compute the correlation between `description_length` and `rating`
- Visualize the relationship with a scatter plot

```python
# Create a new column for the length of product descriptions
df['description_length'] = df['about_product'].fillna('').astype(str).apply(len)

# Correlation with rating
desc_rating_corr = df['description_length'].corr(df['rating'])
print(f"Correlation between description length and rating: {desc_rating_corr:.3f}")
```

### Observation:
- The correlation coefficient is 0.028, which means there is almost no linear relationship between description length and customer rating.
- The scatter plot shows that both short and long descriptions can lead to high or low ratings, with no clear trend or cluster.

### Conclusion:
The length of the product description does not significantly impact customer ratings. A well-written short description can be just as effective as a long one. Instead, the content and clarity of the description likely matter more than its length
